```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
import pandas as pd
```

```python
features = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
target = pd.DataFrame(iris.target,columns=['type'])
```

```python
features[split_num:]#后30%
```

以上测试数据分类存在问题，数据分类不均，下面改进方法，重新划分

```python
from sklearn.model_selection import train_test_split
```

```python
features_train,features_test,target_train,target_test = train_test_split(features,target,test_size =0.3)
```

```python
features_train
```

```python
target_train
```

发现数据成功打乱

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(features_train,target_train)
```

```python
model.predict(features_test)#预测数据
```

```python
target_test.values.flatten()#实际数据
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(target_test.values.flatten(),model.predict(features_test))
```

预测准确率约为96%
